```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import folium
warnings.filterwarnings('ignore')

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.4f}'.format)
```

```python
# DATA_PATH    = "datasets/4g-lte-2024-mobile-signal-measurement-data/4g-lte-2024-mobile-signal-measurement-data.csv"
DATA_PATH    = "datasets/4g-lte-2024-mobile-signal-measurement-data/4g-lte-2024-mobile-signal-measurement-data-clean.csv"
# OUTPUT_PATH  = "datasets/4g-lte-2024-mobile-signal-measurement-data/4g-lte-2024-mobile-signal-measurement-data-clean.csv"
CHUNKSIZE    = 100_000

OPERATORS   = ['3uk', 'ee', 'o2', 'vf']
TOP1_FIELDS = ['pci', 'rsrp', 'rsrq', 'earfcn', 'sinr']
BASE_COLS   = ['latitude', 'longitude', 'month_year']
# can maybe add hour_ref and rnum later

# Only load the columns we actually need
RSRP_COLS   = [f'rsrp_top1_{op}' for op in OPERATORS]
OP_COLS     = [f'{field}_top1_{op}' for op in OPERATORS for field in TOP1_FIELDS]
COLS_TO_KEEP = BASE_COLS + OP_COLS

# Valid ranges for cleaning
RSRP_MIN, RSRP_MAX = -140, -44       # dBm — physically valid 4G RSRP
LAT_MIN,  LAT_MAX  =  49.5,  61.0   # UK bounding box
LON_MIN,  LON_MAX  =  -8.5,   2.0
```

```python
df_raw = pd.read_csv(DATA_PATH, usecols=COLS_TO_KEEP, low_memory=True)
print(f"Loaded: {len(df_raw):,} rows × {df_raw.shape[1]} columns")

# ── 1. Drop rows with no signal from any operator ─────────────────────────────
before = len(df_raw)
df_raw = df_raw[df_raw[RSRP_COLS].notna().any(axis=1)]
print(f"No-coverage rows dropped:    {before - len(df_raw):,}")

# ── 2. GPS sanity check ───────────────────────────────────────────────────────
before = len(df_raw)
df_raw = df_raw[
    df_raw['latitude'].between(49.5, 61.0) &
    df_raw['longitude'].between(-8.5, 2.0)
]
print(f"Out-of-UK GPS rows dropped:  {before - len(df_raw):,}")

# ── 3. Melt wide → long (one row per operator per measurement) ────────────────
records = []
for op in OPERATORS:
    op_df = df_raw[BASE_COLS + [f'{f}_top1_{op}' for f in TOP1_FIELDS]].copy()
    op_df.columns = BASE_COLS + TOP1_FIELDS
    op_df['operator'] = op
    op_df = op_df.dropna(subset=['rsrp'])
    records.append(op_df)

df = pd.concat(records, ignore_index=True)
print(f"After melt:                  {len(df):,} rows")

# ── 4. RSRP range filter ──────────────────────────────────────────────────────
before = len(df)
df = df[df['rsrp'].between(-140, -44)]
print(f"Invalid RSRP rows dropped:   {before - len(df):,}")

# ── 5. Drop rows missing essential columns ────────────────────────────────────
before = len(df)
df = df.dropna(subset=['latitude', 'longitude', 'earfcn', 'pci'])
print(f"Missing essentials dropped:  {before - len(df):,}")

# ── 6. Derive frequency band from EARFCN ─────────────────────────────────────
def earfcn_to_band(e):
    if   0    <= e <= 599:  return 'B1_2100MHz'
    elif 1200 <= e <= 1949: return 'B3_1800MHz'
    elif 2750 <= e <= 3449: return 'B7_2600MHz'
    elif 3600 <= e <= 3799: return 'B8_900MHz'
    elif 6150 <= e <= 6449: return 'B20_800MHz'
    else:                   return 'Other'

df['freq_band'] = df['earfcn'].apply(earfcn_to_band)

# ── 7. Impute rsrq / sinr or drop if too sparse ───────────────────────────────
for col in ['rsrq', 'sinr']:
    null_pct = df[col].isnull().mean() * 100
    if null_pct < 15:
        df[col] = df.groupby('operator')[col].transform(
            lambda x: x.fillna(x.median())
        )
        print(f"{col}: imputed with per-operator median  ({null_pct:.1f}% null)")
    else:
        df = df.drop(columns=[col])
        print(f"{col}: dropped  ({null_pct:.1f}% null — too sparse to impute)")

# ── 8. Collapse near-duplicate GPS readings ───────────────────────────────────
before = len(df)
df['lat_r'] = df['latitude'].round(5)
df['lon_r']  = df['longitude'].round(5)

agg_dict = {'rsrp': 'median', 'pci': 'first',
            'freq_band': 'first', 'month_year': 'first'}
for col in ['rsrq', 'sinr']:
    if col in df.columns:
        agg_dict[col] = 'median'

df = (df.groupby(['lat_r', 'lon_r', 'operator', 'earfcn'], as_index=False)
        .agg(agg_dict)
        .rename(columns={'lat_r': 'latitude', 'lon_r': 'longitude'}))

print(f"Near-duplicates collapsed:   {before - len(df):,}")

# ── Summary ───────────────────────────────────────────────────────────────────
print("\n" + "=" * 45)
print("CLEAN DATASET SUMMARY")
print("=" * 45)
print(f"Rows:        {len(df):,}")
print(f"Columns:     {df.shape[1]}")
print(f"Null values: {df.isnull().sum().sum()}")
print(f"Memory:      {df.memory_usage(deep=True).sum() / 1e6:.1f} MB")
print(f"\nRows per operator:\n{df['operator'].value_counts()}")
print(f"\nRows per frequency band:\n{df['freq_band'].value_counts()}")
print(f"\nRSRP (dBm):\n{df['rsrp'].describe().round(2)}")

# ── Export ────────────────────────────────────────────────────────────────────
df.to_csv(OUTPUT_PATH, index=False)
print(f"\n✓ Saved to {OUTPUT_PATH}")
```
